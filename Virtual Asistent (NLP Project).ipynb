```python
import pyttsx3
import smtplib
import speech_recognition as sr
import pywhatkit
import datetime
import time

# Text to Speech engine
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
engine.setProperty('rate', 185)

def Speak(audio):
    engine.say(audio)
    engine.runAndWait()

def TakeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source, timeout=4)

    try:
        print("Understanding...")
        query = r.recognize_google(audio, language='en-in')
        print(f"Master said: {query}\n")
        return query
    except Exception as e:
        print("Say that again please...")
        return "None"

# Greet when the Program starts
def greetme():
    hour = int(datetime.datetime.now().hour)
    if hour >= 0 and hour < 12:
        Speak('Good Morning Sir')
    elif hour >= 12 and hour < 18:
        Speak('Good Afternoon Sir')
    else:
        Speak('Good Evening Sir')
    Speak('I am Arya. How can I help you, Sir?')

if __name__ == "__main__":
    greetme()  # Greet the user when the program starts

    while True:
        query = TakeCommand().lower()
        
        if "wake up" in query:
            greetme()
        
        elif "youtube" in query:
            text_to_speak = "Okay, I will bring that up on YouTube for you."
            Speak(text_to_speak)
            pywhatkit.playonyt(query)
        
        elif "joke" in query:
            joke_to_speak = "Why don't scientists trust atoms? Because they make up everything!"
            Speak(joke_to_speak)
        
        elif "go to sleep" in query:
            text_to_speak = "Okay, sir you can call me anytime."
            Speak(text_to_speak)
        
        elif "quit" in query or "exit" in query or "stop" in query:
            text_to_speak = "Okay, sir. Shutting down. Have a good day Thank You !"
            Speak(text_to_speak)
            break
        
        # Interview questions added here
        elif "interview" in query or "job" in query:
            Speak("Sure, I can help with that. Let's start with some common interview questions.")
            time.sleep(10)  # Delay for 1 minute
            Speak("Tell me about yourself.")
            time.sleep(30)  # Delay for 1 minute
            Speak("Why are you interested in this position?")
            time.sleep(30)  # Delay for 1 minute
            Speak("What are your strengths and weaknesses?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Can you describe a challenging situation you faced at work and how you handled it?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Where do you see yourself in five years?")
            time.sleep(30)  # Delay for 1 minute
            Speak("What do you know about our company?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Describe a project you successfully completed. What was your role?")
            time.sleep(30)  # Delay for 1 minute
            Speak("How do you prioritize your work?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Give an example of how you work under pressure.")
            time.sleep(30)  # Delay for 1 minute
            Speak("Describe a time you worked as part of a team. What was your role, and what was the outcome?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Explain [specific technical concept or skill related to the job].")
            time.sleep(30)  # Delay for 1 minute
            Speak("How do you stay updated with the latest developments in your field?")
            time.sleep(30)  # Delay for 1 minute
            Speak("Do you have any questions for us?")

        else:
            text_to_speak = "Okay, I will bring that up on Google for you."
            Speak(text_to_speak)
            pywhatkit.search(query)
```
